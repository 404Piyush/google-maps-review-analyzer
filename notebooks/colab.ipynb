# Google Maps Review Analyzer — Colab

Free T4 GPU + Ollama + this repo. End-to-end run in one click.

**Steps:**
1. Click *Runtime → Run all* (Colab spins up a T4 GPU for free)
2. Wait ~3 minutes for Ollama to start and `gemma2:2b` to download
3. The last cell dumps the generated executive report

Don't have a Google Places API key? Use the built-in sample reviews (default).

```python
%%bash
set -e
curl -fsSL https://deb.nodesource.com/setup_20.x | bash - >/dev/null
apt-get install -y nodejs >/dev/null
curl -fsSL https://ollama.com/install.sh | sh >/dev/null
nohup ollama serve >/tmp/ollama.log 2>&1 &
sleep 5
ollama pull gemma2:2b
git clone https://github.com/404Piyush/google-maps-review-analyzer.git
cd google-maps-review-analyzer
npm install --omit=optional --no-audit --no-fund
echo 'Setup complete'
```

```python
%%bash
cd google-maps-review-analyzer
cat > sample-reviews.json <<'EOF'
[{"name":"Demo User","time":"1 day ago","stars":5,"text":"The food was amazing and the staff went above and beyond to make us feel welcome. Highly recommend the chef's special."},
 {"name":"Another User","time":"2 days ago","stars":4,"text":"Great atmosphere and friendly service. Slightly pricey but worth it for the quality."},
 {"name":"Critic","time":"3 days ago","stars":2,"text":"Long wait times and the music was way too loud. Food was mediocre."},
 {"name":"Happy Customer","time":"1 week ago","stars":5,"text":"Best experience I've had at a restaurant in years. Will definitely come back."}]
EOF
mkdir -p output
cp sample-reviews.json output/reviews.json
echo 'Sample loaded'
```

```python
%%bash
cd google-maps-review-analyzer
node topic-analysis.js --model=fast --batch-size=4 --input=output/reviews.json
```

```python
%%bash
cd google-maps-review-analyzer
echo '================================================================'
echo 'Final report:'
echo '================================================================'
cat output/analysis-report.md
```
